## Ejercicio 1: Suma de dos números

Crea una función llamada `sumar(a, b)` que retorne la suma de los parámetros `a` y `b`.

```python
def sumar(a, b):
    ### INICIO DE TU CÓDIGO ###
    return a + b
    ### FIN DE TU CÓDIGO ###
```

```python
assert sumar(2, 3) == 5, "Error: 2 + 3 debe ser 5"
assert sumar(-1, 1) == 0, "Error: -1 + 1 debe ser 0"
print("¡Ejercicio 1 verificado correctamente!")
```

## Ejercicio 2: Verificación de Número Par

Crea una función llamada `es_par(numero)` que devuelva `True` si el número es par y `False` si es impar.

```python
def es_par(numero):
    ### INICIO DE TU CÓDIGO ###
    return numero % 2 == 0
    ### FIN DE TU CÓDIGO ###
```

```python
assert es_par(4) == True, "Error: 4 debe ser par"
assert es_par(7) == False, "Error: 7 debe ser impar"
print("¡Ejercicio 2 verificado correctamente!")
```
