# 🚀 N00 — Bienvenido: tu primer programa hoy mismo

**Módulo 0 «Despegue»** · Semana 1 · ⏱ 90–120 min

Hola. 👋 Esto que estás viendo es un **cuaderno interactivo**: un documento donde el texto y el código conviven, y donde el código **se puede ejecutar aquí mismo**.

En los próximos **10 minutos** vas a ejecutar tu primer programa. Sí, hoy. Sí, tú.

▶️ **Empieza ejecutando la celda de abajo**: haz clic sobre ella y presiona `Shift+Enter` (o el botón ▶ a su izquierda).

````python
# @title ⚙️ Activar mi cuaderno — ejecútame primero { display-mode: "form" }
import ipywidgets as W
from IPython.display import display, HTML

# Paleta del AVA (módulo 0: violeta → azul)
AZUL, VIOLETA, VIOLETA_OSC = "#2a78d6", "#4a3aa7", "#2e2377"
VERDE, ROJO, AMBAR, GRIS = "#0ca30c", "#d03b3b", "#eda100", "#52514e"
FUENTE = "font-family:system-ui,-apple-system,Segoe UI,sans-serif;"

# ── Sistema de progreso ────────────────────────────────────────
XP = {"pts": 0, "meta": 60, "hechos": set()}
_CHECKS = {}
_usadas = {}

_PISTAS = {
    "N1": ["Busca el texto entre comillas y reemplázalo por TU nombre (deja las comillas).",
           "Debe quedar algo como: mi_nombre = \"Carolina\" — con tu nombre entre comillas.",
           "Cambia SOLO lo que está entre comillas y vuelve a ejecutar la celda con Shift+Enter."],
    "R1": ["Python no sabe qué es Hola sin comillas: cree que le hablas de una variable.",
           "El texto siempre va entre comillas: \"Hola\".",
           "La línea debe quedar: mensaje = \"Hola\" — y vuelve a ejecutar."],
    "T1": ["Recuerda qué hiciste con la celda de tu nombre: la ejecutaste TÚ, cuando quisiste.",
           "El cuaderno no ejecuta nada solo: espera a que tú se lo pidas.",
           "Las celdas corren en el orden en que TÚ las ejecutas — por eso conviene ir de arriba hacia abajo."],
    "A1": ["Piensa: ¿qué te deja aprendiendo más, que te den la respuesta o que te ayuden a encontrarla?",
           "El asistente es un tutor, no una máquina de respuestas.",
           "La mejor opción es pedirle que te haga preguntas para descubrir el error tú."],
    "H1": ["Fue un siglo ANTES de que existiera el primer computador electrónico.",
           "Escribió el algoritmo para una máquina que ni siquiera llegó a construirse: la Máquina Analítica.",
           "Ada Lovelace, en 1843."],
}

def _html_barra():
    pct = min(100, round(XP["pts"] * 100 / XP["meta"]))
    med = "🥇" if XP["pts"] >= 60 else "🥈" if XP["pts"] >= 35 else "🥉" if XP["pts"] >= 15 else "⬜"
    return ("<div style='max-width:540px;margin-top:8px;" + FUENTE + "'>"
            "<div style='display:flex;justify-content:space-between;font-size:13px;color:" + GRIS + ";'>"
            "<span>Progreso del cuaderno</span><span><b>" + str(XP["pts"]) + "</b> / " + str(XP["meta"]) + " XP&nbsp;" + med + "</span></div>"
            "<div style='background:#8884;border-radius:99px;height:12px;overflow:hidden;'>"
            "<div style='width:" + str(pct) + "%;height:100%;background:linear-gradient(90deg," + VIOLETA + "," + AZUL + ");'></div>"
            "</div></div>")

def _tarjeta_ok(ej, xp, nuevo):
    extra = "+" + str(xp) + " XP" if nuevo else "(ya lo tenías ganado 😉)"
    return ("<div style='border-left:4px solid " + VERDE + ";padding:8px 12px;" + FUENTE + "'>"
            "<b style='color:" + VERDE + ";'>✅ ¡Conseguido!</b> " + extra + "</div>")

def _tarjeta_no(msg):
    return ("<div style='border-left:4px solid " + ROJO + ";padding:8px 12px;" + FUENTE + "'>"
            "<b style='color:" + ROJO + ";'>🤏 Casi.</b> " + msg +
            "<br><span style='color:" + GRIS + ";font-size:13px;'>Inténtalo de nuevo, no hay penalización. El botón 💡 te da hasta 3 pistas.</span></div>")

def _calificar(ok, ej, xp, msg_no=""):
    if ok:
        nuevo = ej not in XP["hechos"]
        if nuevo:
            XP["hechos"].add(ej); XP["pts"] += xp
        display(HTML(_tarjeta_ok(ej, xp, nuevo) + _html_barra()))
    else:
        display(HTML(_tarjeta_no(msg_no)))

def pista(ej):
    lista = _PISTAS.get(ej, [])
    n = _usadas.get(ej, 0)
    if n >= len(lista):
        display(HTML("<div style='" + FUENTE + "color:" + GRIS + ";'>🙌 Ya usaste las 3 pistas. "
                     "Prueba con el asistente: <i>pídele que te haga preguntas</i>.</div>"))
        return
    _usadas[ej] = n + 1
    display(HTML("<div style='border-left:4px solid " + AMBAR + ";padding:6px 12px;" + FUENTE + "'>"
                 "💡 <b>Pista " + str(n + 1) + " de " + str(len(lista)) + ":</b> " + lista[n] + "</div>"))

def _fila_botones(ej, xp, obtener, msg_no):
    bv = W.Button(description="Verificar ✔", button_style="primary")
    bp = W.Button(description="💡 Pista")
    out = W.Output()
    def _v(_):
        with out:
            out.clear_output()
            _calificar(_CHECKS[ej](obtener()), ej, xp, msg_no)
    def _p(_):
        with out:
            pista(ej)
    bv.on_click(_v); bp.on_click(_p)
    return W.HBox([bv, bp]), out

def quiz(ej, xp, pregunta, opciones, correcta, msg_no=""):
    _CHECKS[ej] = lambda v, c=correcta: v == c
    rb = W.RadioButtons(options=opciones, index=None, layout=W.Layout(width="max-content"))
    fila, out = _fila_botones(ej, xp, lambda: rb.value, msg_no)
    display(W.VBox([W.HTML("<div style='" + FUENTE + "font-size:15px;'><b>" + pregunta + "</b></div>"), rb, fila, out]))

# ── Verificadores de las celdas de código ──────────────────────
def verificar_mi_nombre(nombre):
    ok = (isinstance(nombre, str) and len(nombre.strip()) >= 2
          and "escribe tu nombre" not in nombre.lower())
    _calificar(ok, "N1", 10,
               "Reemplaza el texto entre comillas por tu nombre y vuelve a ejecutar. (Ayuda: pista(\"N1\"))")

def verificar_reparacion(mensaje):
    _calificar(isinstance(mensaje, str) and len(mensaje) > 0, "R1", 10,
               "El texto necesita comillas. (Ayuda: pista(\"R1\"))")

# ── Bienvenida ─────────────────────────────────────────────────
display(HTML(
    "<div style='background:linear-gradient(135deg," + VIOLETA_OSC + "," + AZUL + ");border-radius:14px;"
    "padding:22px 26px;color:#fff;" + FUENTE + "'>"
    "<div style='font-size:12px;letter-spacing:2px;opacity:.85;'>MÓDULO 0 · DESPEGUE · SEMANA 1</div>"
    "<div style='font-size:26px;font-weight:700;margin:6px 0;'>🚀 ¡Acabas de ejecutar tu primera celda!</div>"
    "<div style='font-size:14px;opacity:.92;'>En serio: esto que acaba de pasar —le diste una orden a un "
    "computador y la cumplió— es la esencia de todo lo que viene. El resto del semestre es volverte "
    "cada vez mejor dándole órdenes más interesantes.</div>"
    "<div style='margin-top:12px;font-size:14px;background:#ffffff22;border-radius:8px;padding:8px 12px;'>"
    "🎯 Meta de hoy: tu primer programa, conocer tu AVA y ganar tu primera insignia. Nada de exámenes.</div></div>"))
display(HTML(_html_barra()))
print("⚙️ Cuaderno activado.")
````

````python
# @title 🧭 Mi punto de partida (ejecútame y responde)
_lay = W.Layout(width="max-content")
_lbl = lambda t: W.HTML("<div style='" + FUENTE + "font-size:15px;margin-top:10px;'><b>" + t + "</b></div>")

_dx_exp = W.RadioButtons(options=[
    "Nunca he programado",
    "He probado algo (Scratch, fórmulas de Excel, un tutorial suelto)",
    "He escrito código alguna vez",
    "Programo con cierta comodidad"], index=None, layout=_lay)

_dx_q2 = W.RadioButtons(options=["18", "20", "24", "32"], index=None, layout=_lay)
_dx_q3 = W.RadioButtons(options=["Norte", "Sur", "Este", "Oeste"], index=None, layout=_lay)
_dx_q4 = W.RadioButtons(options=["2", "3", "5", "No estoy seguro"], index=None, layout=_lay)

_dx_conf = W.RadioButtons(options=[
    "1 — Nervios totales 😰",
    "2 — Algo de susto",
    "3 — Ni fu ni fa",
    "4 — Con ganas",
    "5 — ¡Con toda! 🔥"], index=None, layout=_lay)

_dx_btn = W.Button(description="Ver mi punto de partida 🧭", button_style="info")
_dx_out = W.Output()

def _dx_ver(_):
    with _dx_out:
        _dx_out.clear_output()
        if None in (_dx_exp.value, _dx_q2.value, _dx_q3.value, _dx_q4.value, _dx_conf.value):
            display(HTML("<div style='border-left:4px solid " + AMBAR + ";padding:8px 12px;" + FUENTE + "'>"
                         "🤗 Te faltan preguntas por responder. Tranquilo: no hay respuestas malas.</div>"))
            return
        _logica = (_dx_q2.value == "32") + (_dx_q3.value == "Este") + (_dx_q4.value == "5")
        if _logica == 3:
            _msg = "🌟 Traes la intuición lógica afilada. El curso te dará el lenguaje para convertirla en programas."
        elif _logica == 2:
            _msg = "💪 Ya piensas en patrones y secuencias. Lo que falta es justo lo que vamos a entrenar."
        else:
            _msg = "🌱 Estás exactamente donde este curso está diseñado para empezar. Esto se entrena — y para eso estamos."
        _nervios = _dx_conf.value.startswith(("1", "2"))
        _conf_msg = ("Los nervios del arranque son normales — buena parte de tus compañeros llega igual. "
                     "Este AVA avanza a tu ritmo: intentos ilimitados, pistas y un tutor 24/7."
                     if _nervios else "Esa actitud es medio curso ganado. 😎")
        nuevo = "DX" not in XP["hechos"]
        if nuevo:
            XP["hechos"].add("DX"); XP["pts"] += 10
        display(HTML(
            "<div style='border:2px solid " + VIOLETA + ";border-radius:12px;padding:14px 18px;max-width:560px;" + FUENTE + "'>"
            "<div style='font-size:17px;font-weight:700;color:" + VIOLETA + ";'>🧭 Tu punto de partida</div>"
            "<div style='margin-top:8px;font-size:14px;'>"
            "<b>Experiencia:</b> " + _dx_exp.value + "<br>"
            "<b>Intuición lógica:</b> " + _msg + "<br>"
            "<b>Ánimo:</b> " + _conf_msg + "</div>"
            "<div style='margin-top:10px;font-size:12.5px;color:" + GRIS + ";'>Esta foto es tuya. No es una nota "
            "ni una etiqueta: en la semana 16 volverás a mirarla y no te vas a reconocer. 📈</div></div>"
            + ("" if not nuevo else _html_barra())))

_dx_btn.on_click(_dx_ver)
display(W.VBox([
    _lbl("1 · ¿Cuál frase te describe mejor hoy?"), _dx_exp,
    _lbl("2 · ¿Qué número sigue? &nbsp;2, 4, 8, 16, ___"), _dx_q2,
    _lbl("3 · Un robot mira al Norte. Gira a la derecha, gira a la derecha, gira a la izquierda. ¿Hacia dónde mira?"), _dx_q3,
    _lbl("4 · Sigue estos pasos en orden: x = 3, luego x = x + 2. ¿Cuánto vale x al final?"), _dx_q4,
    _lbl("5 · ¿Cómo llegas anímicamente a este curso?"), _dx_conf,
    W.HTML("<div style='height:8px;'></div>"), _dx_btn, _dx_out]))
````

Marca **solo lo que de verdad hiciste hoy** y reclama tu insignia de despegue:

````python
# @title 🚀 Reclamar mi insignia (ejecútame y marca lo que lograste)
_ck = [W.Checkbox(description=d, indent=False, layout=W.Layout(width="max-content")) for d in [
    "Ejecuté y modifiqué mi primer programa",
    "Sé moverme por un cuaderno (celdas, Shift+Enter, orden)",
    "Rompí algo a propósito y leí el error sin pánico",
    "Sé pedirle ayuda al asistente sin pedirle la respuesta",
    "Completé mi punto de partida"]]
_bc = W.Button(description="Reclamar insignia 🚀", button_style="success")
_oc = W.Output()

def _cerrar(_):
    with _oc:
        _oc.clear_output()
        faltan = sum(not c.value for c in _ck)
        if faltan:
            display(HTML("<div style='border-left:4px solid " + AMBAR + ";padding:8px 12px;" + FUENTE + "'>"
                         "🤔 Te quedan " + str(faltan) + " punto(s) sin marcar. Vuelve a esa sección "
                         "o pregúntale al asistente — hoy no hay afán.</div>"))
        elif XP["pts"] >= XP["meta"]:
            display(HTML(
                "<div style='background:linear-gradient(135deg,#2e2377,#2a78d6);border-radius:14px;"
                "padding:20px 24px;color:#fff;" + FUENTE + "max-width:560px;'>"
                "<div style='font-size:24px;font-weight:700;'>🚀 Insignia: Despegue</div>"
                "<div style='font-size:14px;margin-top:6px;opacity:.95;'>Hoy ejecutaste tu primer programa, "
                "leíste tu primer error, conociste a tu tutor y trazaste tu punto de partida: " + str(XP["pts"]) +
                " XP. Oficialmente, despegaste. 🛫<br><br><b>Próxima parada → N01: De problemas a algoritmos</b> "
                "(la semana que viene pensamos como programadores... sin tocar el teclado).</div></div>"))
        else:
            display(HTML("<div style='border-left:4px solid " + AZUL + ";padding:8px 12px;" + FUENTE + "'>"
                         "💪 Llevas <b>" + str(XP["pts"]) + " XP de " + str(XP["meta"]) + "</b>. Revisa qué "
                         "verificador te falta (busca las celdas con ✔) y la insignia es tuya.</div>"))

_bc.on_click(_cerrar)
display(W.VBox(_ck + [_bc, _oc]))
````
